Contexto:
Un estudiante de preparatoria que está cursando la materia de matemáticas necesita resolver ejercicios de geometría relacionados con el rectángulo. Para facilitar sus cálculos y verificar sus resultados, requiere desarrollar un programa en Python que le permita calcular de manera rápida y precisa el área y el perímetro de cualquier rectángulo.


# Fórmulas a utilizar:
Área: A = base × altura
Perímetro: P = 2 × (base + altura)

![texto del vínculo](https://i.ytimg.com/vi/BY0Cea_y04s/maxresdefault.jpg)

```python


def menu():
    print("\n" + "="*70)
    print(" " * 22 + "MENÚ PRINCIPAL" + " " * 22)
    print("="*70)
    print("  🟦 OPCIÓN 1: Calcular el área del rectángulo")
    print("     → Fórmula: A = base × altura")
    print("     → Unidades: cuadradas")
    print("-"*70)
    print("  🟩 OPCIÓN 2: Calcular el perímetro del rectángulo")
    print("     → Fórmula: P = 2 × (base + altura)")
    print("     → Unidades: lineales")
    print("-"*70)
    print("  ⚫ OPCIÓN 3: Salir del programa")
    print("     → Terminar la ejecución del sistema")
    print("="*70)



def area_rectangulo():
    print("\n" + "🟦"*35)
    print(" " * 13 + "CÁLCULO DEL ÁREA DEL RECTÁNGULO")
    print("🟦"*35)
    print("\n 📝 Instrucciones:")
    print("    Ingresa la base y la altura del rectángulo.")
    print("    • Base: lado horizontal del rectángulo")
    print("    • Altura: lado vertical del rectángulo")

    # Solicitar la base al usuario
    base = float(input("\n 📏 Ingresa el valor de la base: "))

    # Solicitar la altura al usuario
    altura = float(input(" 📏 Ingresa el valor de la altura: "))

    # Calcular el área usando la fórmula A = b × h
    area = base * altura

    # Mostrar resultados
    print("\n" + "✨"*35)
    print(" " * 16 + "RESULTADOS")
    print("✨"*35)
    print(f"  📐 Base ingresada:        {base}")
    print(f"  📐 Altura ingresada:      {altura}")
    print(f"  🎯 Área calculada:        {area:.2f}")
    print(f"  📏 Unidades:              cuadradas")
    print(f"  🧮 Fórmula aplicada:      A = base × altura")
    print(f"  🔢 Cálculo:               A = {base} × {altura}")
    print(f"  ✅ Resultado:             A = {area:.2f}")
    print("✨"*35)
    print("\n 💡 Interpretación:")
    print(f"     El rectángulo tiene una superficie total de {area:.2f}")
    print(f"     unidades cuadradas.")



def perimetro_rectangulo():
    print("\n" + "🟩"*35)
    print(" " * 10 + "CÁLCULO DEL PERÍMETRO DEL RECTÁNGULO")
    print("🟩"*35)
    print("\n 📝 Instrucciones:")
    print("    Ingresa la base y la altura del rectángulo.")
    print("    • Base: lado horizontal del rectángulo")
    print("    • Altura: lado vertical del rectángulo")

    # Solicitar la base al usuario
    base = float(input("\n 📏 Ingresa el valor de la base: "))

    # Solicitar la altura al usuario
    altura = float(input(" 📏 Ingresa el valor de la altura: "))

    # Calcular el perímetro usando la fórmula P = 2 × (b + h)
    perimetro = 2 * (base + altura)

    # Mostrar resultados
    print("\n" + "✨"*35)
    print(" " * 16 + "RESULTADOS")
    print("✨"*35)
    print(f"  📐 Base ingresada:        {base}")
    print(f"  📐 Altura ingresada:      {altura}")
    print(f"  🎯 Perímetro calculado:   {perimetro:.2f}")
    print(f"  📏 Unidades:              lineales")
    print(f"  🧮 Fórmula aplicada:      P = 2 × (base + altura)")
    print(f"  🔢 Cálculo:               P = 2 × ({base} + {altura})")
    print(f"  🔢 Cálculo:               P = 2 × {base + altura}")
    print(f"  ✅ Resultado:             P = {perimetro:.2f}")
    print("✨"*35)
    print("\n 💡 Interpretación:")
    print(f"     La distancia total alrededor del rectángulo es de")
    print(f"     {perimetro:.2f} unidades lineales.")

# ============================================================================
# ============================================================================
# 🚀 PROGRAMA PRINCIPAL - PUNTO DE ENTRADA
#
# ============================================================================
# ============================================================================

# Mensaje de bienvenida inicial
print("\n" + "🌟"*70)
print("🌟" + " "*68 + "🌟")
print("🌟" + " " * 13 + "🎓 SISTEMA GEOMÉTRICO EDUCATIVO 🎓" + " " * 13 + "🌟")
print("🌟" + " " * 10 + "📐 CÁLCULO DE RECTÁNGULOS 📐" + " " * 10 + "🌟")
print("🌟" + " " * 16 + "Nivel Preparatoria - Matemáticas" + " " * 16 + "🌟")
print("🌟" + " "*68 + "🌟")
print("🌟"*70)
print("\n 📚 Bienvenido estudiante:")
print("    Este programa te ayudará a calcular:")
print("    • Área de rectángulos (superficie)")
print("    • Perímetro de rectángulos (contorno)")
print("\n 💡 Instrucciones de uso:")
print("    1. Selecciona una opción del menú")
print("    2. Ingresa la base y altura cuando se solicite")
print("    3. Observa los resultados y su interpretación")
print("    4. Elige la opción 3 cuando desees salir")
print("\n 🎯 Objetivo de aprendizaje:")
print("    Comprender y aplicar las fórmulas del rectángulo")

# Inicialización de la variable de control del menú
opcion = 0

# Ciclo principal del programa - se ejecuta mientras no se elija salir
while opcion != 3:
    # Llamar a la función que muestra el menú
    menu()

    # Solicitar al usuario que seleccione una opción
    opcion = int(input("\n 💡 Ingresa el número de la opción deseada (1-3): "))

    # Estructura condicional para ejecutar la función correspondiente
    if opcion == 1:
        # Opción 1: Calcular área del rectángulo
        print("\n 🔄 Calculando área del rectángulo...")
        area_rectangulo()

    elif opcion == 2:
        # Opción 2: Calcular perímetro del rectángulo
        print("\n 🔄 Calculando perímetro del rectángulo...")
        perimetro_rectangulo()

    elif opcion == 3:
        # Opción 3: Salir del programa
        print("\n" + "👋"*70)
        print("👋" + " "*68 + "👋")
        print("👋" + " " * 18 + "¡GRACIAS POR USAR EL PROGRAMA!" + " " * 17 + "👋")
        print("👋" + " " * 23 + "🎉 Éxito en tus estudios 🎉" + " " * 22 + "👋")
        print("👋" + " " * 20 + "📚 ¡Sigue practicando! 📚" + " " * 19 + "👋")
        print("👋" + " " * 15 + "💪 ¡Tú puedes dominar las matemáticas! 💪" + " " * 13 + "👋")
        print("👋" + " "*68 + "👋")
        print("👋"*70 + "\n")

    else:
        # Opción no válida - mostrar mensaje de error
        print("\n" + "⚠️"*35)
        print("⚠️" + " " * 13 + "ATENCIÓN: Opción no válida" + " " * 12 + "⚠️")
        print("⚠️"*35)
        print("\n ❌ Error: La opción ingresada no es correcta.")
        print(" 💡 Sugerencia:")
        print("    • Revisa el menú mostrado arriba")
        print("    • Selecciona un número del 1 al 3")
        print("    • Presiona Enter después de elegir")
        print("    • Intenta de nuevo")
        print("⚠️"*35 + "\n")

```
